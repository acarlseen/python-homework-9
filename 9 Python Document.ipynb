```python
url = 'http://ergast.com/api/f1/2008/5/driverStandings.json'
```

```python
#response object

response = requests.get(url)
print(response)
```

```python
#check and make sure we got a successful response from the API
response.ok
```

```python
#view the response as a JSON
response.json()
```

```python
type(response.json())
```

```python
# We only want the Driver Standings
my_racer_data = response.json()['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings']
my_racer_data
```

```python
def get_racer_info(data):
    new_data = []
    for racer in data:
        racer_dict = {}
        racer_name = f"{racer['Driver']['givenName']} {racer['Driver']['familyName']}"
        
        racer_dict[racer_name] = {
            'first_name' : racer['Driver']['givenName'],
            'last_name' : racer['Driver']['familyName'],
            'position' : racer['position'],
            'wins' : racer['wins'],
            'DOB' : racer['Driver']['dateOfBirth'],
            'nationality': racer['Driver']['nationality'],
            'constructors' : racer['Constructors'][0]['name']
        }
        new_data.append(racer_dict)
    return new_data

get_racer_info(my_racer_data)
```

```python
def get_driver_info_year_and_round(year, r):
    url = f'http://ergast.com/api/f1/{year}/{r}/driverStandings.json'
    response = requests.get(url)
    if not response.json()['MRData']['StandingsTable']['StandingsLists']:
        return 'We had an error loading your data'
    data = response.json()['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings']
    return(get_racer_info(data))

get_driver_info_year_and_round(2019, 4)
```

```python
'''
pokedex = {bulbasaur:
            {
            'ability1': xxxxx
            'ability2': 
            'ability3': etc...
            'base_experience':
            'sprite_front_shiny':
            'hp': base_stat
            'attack': base_stat
            'defense': base_stat}}
'''

url = 'https://pokeapi.co/api/v2/pokemon/'
response = requests.get(url)
print(response)
pokedex = {}

def get_pokedex_info(url):
    dex = url.json()
    i = 1
    result = {}
    for i, able in enumerate(dex['abilities']):
        result[f'ability{i}'] = able['ability']['name']
    result.update({
        'base experience' : dex['base_experience'],
        'sprite' : dex['sprites']['front_shiny'],
        'hp' : dex['stats'][0]['base_stat'],
        'attack' : dex['stats'][1]['base_stat'],
        'defense' : dex['stats'][2]['base_stat']   
    })
    return result
    
    
for pokemon in response.json()['results']:
    pokedex[pokemon['name']] = get_pokedex_info(requests.get(pokemon['url']))

for items in pokedex.items():
    print(items)
    
```
